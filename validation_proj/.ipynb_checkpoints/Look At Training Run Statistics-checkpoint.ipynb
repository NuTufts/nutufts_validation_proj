```python
# Import useful packages
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Import log from training
log = pd.read_csv("train_log-0000000.csv")

# Look at the log
log
```

```python
# What are all the things that were logged?
log.keys()
```

```python
# Make a mask to show data just at snapshots
snapshot_indices_mask = np.full(len(log['epoch']), False)
for snapshot_index in range(9999, 760000, 10000): # 9999, 19999, 29999, etc.
    snapshot_indices_mask[snapshot_index] = True
```

```python
# plot_from_log
# Plot two statistics from the log against each other
# Params:
#   log - The pandas dataframe of the training log
#   xkey - A string. The key of the data to plot as an x axis
#   ykey - A string. The key of the data to plot as a y axis
#   kind_of_plot - A matplotlib plotting function. Default plt.plot. Could use plt.scatter for example
#   data_min_index - An int. The index to start plotting data from. Default 0. Start at 0th item of xkey and ykey array
#   data_max_index - An int. The index to stop plotting data at. Default None. End at last item in xkey and ykey array 
# Returns:
# Note: Makes a matplotlib plot
def plot_from_log(log, xkey, ykey, kind_of_plot=plt.plot, mask=None, data_min_index=0, data_max_index=None):
    # Get data. Apply mask if there is one, otherwise use indices
    xdata = log[xkey][mask] if (not (type(mask) == type(None))) else log[xkey][data_min_index:data_max_index]
    ydata = log[ykey][mask] if (not (type(mask) == type(None))) else log[ykey][data_min_index:data_max_index]
    
    # Plot using the right kind of plot
    kind_of_plot(xdata, ydata)
    
    # Apply labels & title
    plt.xlabel(xkey)
    plt.ylabel(ykey)
    plt.title(f"{ykey} vs. {xkey}")
```

```python
# Look at segmentation accuracy at every epoch we have a snapshot of
plot_from_log(log, "epoch", 'segmentation_accuracy', kind_of_plot=plt.plot, mask=snapshot_indices_mask, data_min_index=0, data_max_index=None)
# plt.xscale('log')
# plt.yscale('log')
plt.loglog()
plt.xlabel("Epoch At Checkpoint")
plt.title("segmentation_accuracy vs. epoch\nat checkpoints")
plt.show()
```
